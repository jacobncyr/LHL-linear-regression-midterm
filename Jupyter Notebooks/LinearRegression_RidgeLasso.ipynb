```python
import os
import numpy as np
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

directory = "../data"
file = os.path.join(directory, "fullgen_tophalf_pokemon.csv")
data = pd.read_csv(file)

# Select relevant columns
columns = ['type_1','type_2', 'totalstat', 'generation', 'avg_power']
data = data[columns]

# Fill missing values in 'type_2' column with a default value (e.g., 'None')
data['type_2'].fillna('None', inplace=True)

# Perform one-hot encoding on the 'type_1' and 'type_2' columns
type_1_encoded = pd.get_dummies(data['type_1'], prefix='type')
type_2_encoded = pd.get_dummies(data['type_2'], prefix='type2')
data_encoded = pd.concat([data, type_1_encoded, type_2_encoded], axis=1)

# Select the relevant encoded columns and other numerical columns
X = data_encoded[['type_Bug', 'type_Dragon', 'type_Electric', 'type_Fairy', 'type_Fighting',
                  'type_Fire', 'type_Flying', 'type_Ghost', 'type_Grass', 'type_Ground',
                  'type_Ice', 'type_Normal', 'type_Poison', 'type_Psychic', 'type_Rock',
                  'type_Steel', 'type_Water',
                  'type2_None', 'type2_Bug', 'type2_Dragon', 'type2_Electric', 'type2_Fairy', 'type2_Fighting',
                  'type2_Fire', 'type2_Flying', 'type2_Ghost', 'type2_Grass', 'type2_Ground',
                  'type2_Ice', 'type2_Normal', 'type2_Poison', 'type2_Psychic', 'type2_Rock',
                  'type2_Steel', 'type2_Water']]

# Convert boolean values to integers
X = X.astype(int)

y = data_encoded['totalstat']
X = sm.add_constant(X)

# Fit Lasso model
lasso_model = sm.OLS(y, X)
lasso_results = lasso_model.fit_regularized(method='elastic_net', alpha=1, L1_wt=1)

# Fit Ridge model
ridge_model = sm.OLS(y, X)
ridge_results = ridge_model.fit_regularized(method='elastic_net', alpha=1, L1_wt=0)

# Get Lasso coefficients and features
lasso_coefficients = lasso_results.params[1:]
lasso_features = X.columns[1:]

# Get Ridge coefficients and features
ridge_coefficients = ridge_results.params[1:]
ridge_features = X.columns[1:]

# Sort Lasso coefficients and features by magnitude
lasso_sorted_indices = np.abs(lasso_coefficients).argsort()[::-1]
lasso_coefficients = lasso_coefficients[lasso_sorted_indices]
lasso_features = lasso_features[lasso_sorted_indices]

# Sort Ridge coefficients and features by magnitude
ridge_sorted_indices = np.abs(ridge_coefficients).argsort()[::-1]
ridge_coefficients = ridge_coefficients[ridge_sorted_indices]
ridge_features = ridge_features[ridge_sorted_indices]

# Select the top 6 Lasso results
top_lasso_coefficients = lasso_coefficients[:6]
top_lasso_features = lasso_features[:6]

# Select the bottom two-thirds of Ridge results
ridge_count = int(len(ridge_coefficients) * 2 / 3)
bottom_ridge_coefficients = ridge_coefficients[ridge_count:]
bottom_ridge_features = ridge_features[ridge_count:]

# Create a figure with subplots
fig, axes = plt.subplots(nrows=2, figsize=(10, 8))

# Plot top 6 Lasso coefficients
sns.barplot(x=top_lasso_coefficients, y=top_lasso_features, ax=axes[0], palette='viridis')
axes[0].set_xlabel('Coefficient', fontsize=10)
axes[0].set_ylabel('Feature', fontsize=10)
axes[0].set_title('Top 6 Lasso Coefficients', fontsize=12)

# Plot bottom two-thirds of Ridge coefficients
sns.barplot(x=bottom_ridge_coefficients, y=bottom_ridge_features, ax=axes[1], palette='viridis')
axes[1].set_xlabel('Coefficient', fontsize=10)
axes[1].set_ylabel('Feature', fontsize=10)
axes[1].set_title('Bottom Two-Thirds of Ridge Coefficients', fontsize=12)

plt.tight_layout()
plt.show()

```
